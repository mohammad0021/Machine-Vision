```python
import cv2 as cv
import numpy as np
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
```

```python
img= cv.imread("images/m7.jpg")
img= cv.cvtColor(img, cv.COLOR_BGR2RGB)
imshow(img)
```

```python
kernel= np.array([[0,1,0],[1,-4,1], [0,1,0]])
img_edges= cv.filter2D(img, 0, kernel)
imshow(img_edges);
```

```python
img_gaussian= cv.GaussianBlur(img, (5,5), 2.0)
unsharp_image = cv.addWeighted(img_gaussian, 1, img_edges , 1, 0, img)

plt.figure(figsize=(20,10))
ax= plt.subplot(1,2,1)
ax.imshow(img_gaussian);

ax2= plt.subplot(1,2,2)
ax2.imshow(unsharp_image);
```

```python
img_medianblur= cv.medianBlur(img, 1)
unsharp_image = cv.addWeighted(img_medianblur, 1, img_edges , 1, 0, img)

plt.figure(figsize=(20,10))
ax= plt.subplot(1,2,1)
ax.imshow(img_medianblur);

ax2= plt.subplot(1,2,2)
ax2.imshow(unsharp_image);
```
